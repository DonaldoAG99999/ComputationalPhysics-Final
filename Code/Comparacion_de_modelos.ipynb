```python
import numpy as np
import plotly.graph_objects as go
import pandas as pd
from plotly.subplots import make_subplots
```

```python
fig = make_subplots(1,2,x_title='# Epochs',subplot_titles=('Accuracy', 'Loss'))

nep = 50

data = pd.read_csv('acc_loss_df.csv')
data2 = pd.read_csv('acc_loss_fft_df5.csv')
data3 = pd.read_csv('acc_loss_fft_df3.csv')
acc = data['Accuracy']
acc2 = data2['Accuracy']
acc3 = data3['Accuracy']
loss = data['Loss']
loss2 = data2['Loss']
loss3 = data3['Loss']



vec = np.linspace(0, nep, 50)

fig.add_trace(go.Scatter(x = vec,y = acc, name='Accuracy_normal_model'),1,1)
fig.add_trace(go.Scatter(x = vec,y = acc2, name='Accuracy_FFT(real)_model'),1,1)
fig.add_trace(go.Scatter(x = vec,y = acc3, name='Accuracy_FFT(abs)_model'),1,1)


fig.add_trace(go.Scatter(x = vec,y = loss, name='Loss_normal_model'),1,2)
fig.add_trace(go.Scatter(x = vec,y = loss2, name='Loss_FFT(real)_model'),1,2)
fig.add_trace(go.Scatter(x = vec,y = loss3, name='Loss_FFT(abs)_model'),1,2)
fig.update_layout(title='ANÁLISIS DE MODELO', font_size=12)


fig.show()
```
